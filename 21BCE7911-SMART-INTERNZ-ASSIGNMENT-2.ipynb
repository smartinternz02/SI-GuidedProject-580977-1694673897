```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('car_crashes')
df
```

```python
df.info()
```

```python
sns.scatterplot(x='speeding',y='alcohol',data=df)
```

```python
sns.scatterplot(x='speeding',y='not_distracted',data=df)
```

```python
sns.lineplot(x='alcohol',y='speeding',data=df,ci=None)
```

```python
sns.distplot(df['alcohol'])
```

```python
sns.relplot(x='speeding',y='alcohol',data=df,hue='not_distracted')
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
car=fig.add_axes([0.1,0.1,0.8,0.8])
sns.barplot(data=df,x='alcohol',y='speeding',ci=None)
```

```python
sns.jointplot(x='not_distracted',y='speeding',data=df)
```

```python
sns.boxplot(x='total',y='not_distracted',data=df)
```

```python
corr=df.corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='YlGnBu')
```
